```python
tables = [{'topic(member)' : 1, 'table' : 'AG.LND.AGRI.ZS', 'info' :	'Agricultural land (% of land area)', 'class' : 'feature', 'format' : 'percent'}, 
          {'topic(member)' : 1, 'table' : 'AG.AGR.TRAC.NO', 'info' : 'Agricultural machinery, tractors', 'class' : 'feature', 'format' : 'gross'},
          {'topic(member)' : 3, 'table' : 'NY.GDP.MKTP.KD.ZG', 'info' : 'GDP growth (annual %)', 'class' : 'feature', 'format' : 'percent'},
          {'topic(member)' : 3, 'table' : 'NY.GDP.PCAP.PP.KD', 'info': 'GDP per capita, PPP (constant 2017 international $)', 'class' : 'feature', 'format' : 'gross'},
          {'topic(member)' : 5, 'table' : 'EG.USE.COMM.FO.ZS', 'info' : 'Fossil fuel energy consumption', 'class' : 'feature', 'format' : 'gross'},
          {'topic(member)' : 5, 'table' : 'EG.USE.COMM.CL.ZS', 'info': 'Alternative and nuclear energy (% of total energy use)', 'class' : 'feature', 'format' : 'percent'},
          {'topic(member)' : 6, 'table' : 'EG.ELC.FOSL.ZS', 'info' : 'Electricity production from oil, gas and coal sources (% of total)', 'class' : 'feature', 'format' : 'percent'},
          {'topic(member)' : 6, 'table' : 'AG.LND.FRST.ZS', 'info' : 'Forest area (% of land area)', 'class' : 'feature', 'format' : 'percent'},
          {'topic(member)' : 19, 'table' : 'EN.ATM.CO2E.PC', 'info' : 'CO2 emissions (metric tons per capita)', 'class' : 'label', 'format' : 'gross'}]
tables_df = pd.DataFrame(tables)
tables_df
```

```python
for sheet in tables:
    print(sheet)
    df_save = wb.data.DataFrame(sheet['table'], list(wb.economy.Series().index), time = range(1950,2024))
    topic_name = topics_df.loc[topics_df['id'] == str(sheet['topic(member)'])]['dir'].values[0]

    df_save.to_csv(f'data/{topic_name}_{sheet['table']}.csv')
```
